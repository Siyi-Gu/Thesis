### NHTSA Data Process

```python
import pandas as pd
```

```python
fields = pd.read_excel('fields.xlsx')
fields
```

```python
col_names = fields['Name'].tolist()
col_names
```

```python
data = pd.read_csv('COMPLAINTS_RECEIVED_2020-2023.txt', sep="\t", header=None)
data
```

```python
data.columns = col_names
#data = pd.DataFrame(data, columns= col_names)
data
```

```python
com_des_df = data[['COMPDESC', 'CDESCR']]
com_des_df
```

```python
com_des_df = com_des_df.drop_duplicates()
com_des_df
```

```python
com_des_df = com_des_df.apply(lambda x: x.astype(str).str.lower()).sample(n=100)
com_des_df
```

```python
from sklearn.model_selection import train_test_split
train, test = train_test_split(com_des_df, test_size=0.2)
```

```python
train['CDESCR'].to_csv('train.txt', sep="\n", index=False, header=None)
```

```python
test
```

```python
test.to_csv('test.csv', index=False, header=['sent1', 'sent0'])
```
